### Question- 4 Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix.



 Note: When the record/entry has "M"(indicates millions) as suffix you
need to multiply that value with 1000000
 When the record/entry has "K"(indicates thousands) as suffix you
need to multiply that value with 1000

```python
Unique_Data = FiFa_df[['Value', 'Wage','Release Clause']]
Unique_Data.sample(10)
```

```python
Unique_Data.dtypes
```

```python
def clean(x):
    if isinstance(x,str):
        return(x.replace('€','').replace('.','').replace('M','000000').replace('K','000'))
    return(x)
```

```python
FiFa_df['Value'] = FiFa_df['Value'].apply(clean).astype('float64')
FiFa_df['Wage'] = FiFa_df['Wage'].apply(clean).astype('float64')
FiFa_df['Release Clause'] = FiFa_df['Release Clause'].apply(clean).astype('float64')
```

### Question-5  Convert the column "Joined" into integer data type with keeping only the year. 

```python
FiFa_df['Joined'].head(7)
```

```python
def convert_joined(obj):
    
    if isinstance(obj, str):
        
        year_part = obj.split(',')[1].split()
        numeric_value = pd.to_numeric(year_part, errors='coerce')
        return numeric_value
    return obj
```

```python
FiFa_df['Joined'] = FiFa_df['Joined'].apply(convert_joined).replace(np.nan, 0).astype(int)
```

```python
FiFa_df['Joined'].head(7)
```

### Question-18  Find the position in the pitch where the maximum number of players play and the position where the minimum number of players play? Display it using a plot. 

```python
position_count = FiFa_df['Position'].value_counts()
position_count
```

### Inferences: 

 The maximum number of players play are in the position 'ST' (striker) and 'LF' (Left Forward) around 2000 and 15 players.
